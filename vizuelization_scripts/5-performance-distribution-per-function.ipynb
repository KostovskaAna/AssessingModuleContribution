```python
import pandas as pd
import numpy as np
import plotly.figure_factory as ff
from plotly.subplots import make_subplots



dim=5
budget=2000
df = pd.read_csv('../data/processed_data_modCMA/conf_perf2vec.csv', index_col=0)
df_grid = pd.read_csv('../data/raw_data/dt_grid_v2.csv', index_col=0)
variable = f'.*_{budget}_{dim}'           
df_sub = df[df.index.str.contains(fr'\b{variable}\b', regex=True)]
df_sub.index = [ int(i.split("_")[0]) for i in df_sub.index]
    
fig = make_subplots(rows=2, cols=5, vertical_spacing=0.015, shared_yaxes=True, shared_xaxes = True, horizontal_spacing=0.01)
        

# module mirrored


g1 = df_sub.iloc[:,0:5]
g2 = df_sub.iloc[:,5:9]
g3 = df_sub.iloc[:,9:14]
g4 = df_sub.iloc[:,14:19]
g5 = df_sub.iloc[:,19:24]

groups = [g1, g2, g3, g4, g5]

row = 1
for module in ['mirrored', 'local_restart']:
    col = 1
    for df_g in groups:
        module_values = df_grid[module]
        df_g = df_g.join(module_values)
        df_g = df_g.replace({np.nan: 'None'})
        unique_module_values = np.unique(df_g[module])
        hist_data = []
        for m in range(0, len(unique_module_values)):
            unique_module_value = unique_module_values[m]
            df_sub_m = df_g[df_g[module]==unique_module_value]
            df_sub_m = df_sub_m.drop([module], axis=1)
            dist = df_sub_m.mean(axis=1).values
            dist = [np.log10(10**(-8)) if d == 0 else np.log10(d) for d in dist]
            hist_data.append(dist)

        subfig = ff.create_distplot(hist_data, ['m1', 'm2', 'm3'], show_hist=False, show_rug=False)
        fig.add_trace(subfig['data'][0], row = row, col=col)
        fig.add_trace(subfig['data'][1], row = row, col=col)
        fig.add_trace(subfig['data'][2], row = row, col=col)
        col+=1
    row +=1


fig.update_layout(
    autosize=False,
    width=700,
    height=500,
    showlegend=False,
    margin=dict(l=0,r=0,b=0,t=0),
)
fig['layout']['yaxis']['title']='mirrored'
fig['layout']['yaxis6']['title']='local restart'
fig['layout']['xaxis6']['title']='Separable'
fig['layout']['xaxis7']['title']='Moderate'
fig['layout']['xaxis8']['title']='Ill-conditioned'
fig['layout']['xaxis9']['title']='Multi-modal'
fig['layout']['xaxis10']['title']='Weakly-structured'
fig.show()
```
